```python
import random
import string
import json
import requests
import time
import psutil
import csv
import matplotlib.pyplot as plt
```

```python
# Configuracin
fhir_server_url = "http://localhost:8080/fhir"
num_patients = 100

# Listas para almacenar datos
tiempos_respuesta = []
consumo_memoria = []

# Funcin para generar un paciente FHIR aleatorio (sin identifier)
def generar_paciente():
    return {
        "resourceType": "Patient",
        "name": [
            {
                "use": "official",
                "family": "".join(random.choices(string.ascii_letters, k=10)),  # Apellido aleatorio
                "given": ["".join(random.choices(string.ascii_letters, k=8))]  # Nombre aleatorio
            }
        ],
        "birthDate": f"{random.randint(1940, 2004):04d}-{random.randint(1, 12):02d}-{random.randint(1, 28):02d}"  # Fecha de nacimiento aleatoria
    }

# Funcin para cargar un recurso FHIR, medir el tiempo y registrar el consumo de memoria
def cargar_recurso(recurso):
    start_time = time.perf_counter()
    headers = {"Content-Type": "application/fhir+json", "Authorization": "Bearer 12345"}  # Asegrate de usar tu token Bearer real
    response = requests.post(fhir_server_url + "/" + recurso["resourceType"], headers=headers, data=json.dumps(recurso))
    end_time = time.perf_counter()

    # Registrar tiempo de respuesta
    tiempo_respuesta = end_time - start_time
    tiempos_respuesta.append(tiempo_respuesta)

    # Registrar consumo de memoria
    process = psutil.Process()
    memory_info = process.memory_info()
    consumo_memoria.append(memory_info.rss / (1024 * 1024))  # Convertir a MB

# Generacin y carga de pacientes
for _ in range(num_patients):
    paciente = generar_paciente()
    cargar_recurso(paciente)

# Guardar tiempos de respuesta en un archivo CSV
csv_file = "tiempos_respuesta.csv"
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Numero de paciente', 'Tiempo de respuesta (segundos)'])
    for i, tiempo in enumerate(tiempos_respuesta):
        writer.writerow([i + 1, tiempo])
    
print(f"Tiempos de respuesta guardados en {csv_file}")

# Anlisis de resultados (opcional)
plt.figure(figsize=(12, 6))

# Grfico del tiempo de respuesta
plt.subplot(2, 1, 1)
plt.plot(range(1, num_patients + 1), tiempos_respuesta, label="Tiempo de respuesta (segundos)")
plt.xlabel("Nmero de pacientes")
plt.ylabel("Tiempo de respuesta (segundos)")
plt.legend()
plt.title("Tiempo de respuesta del servidor")

# Grfico del consumo de memoria
plt.subplot(2, 1, 2)
plt.plot(range(1, num_patients + 1), consumo_memoria, label="Consumo de memoria (MB)")
plt.xlabel("Nmero de pacientes")
plt.ylabel("Consumo de memoria (MB)")
plt.legend()
plt.title("Consumo de memoria durante la carga de pacientes")

plt.tight_layout()
plt.show()
```
